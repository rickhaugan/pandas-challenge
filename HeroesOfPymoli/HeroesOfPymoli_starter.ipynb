```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items_count = purchase_data["Item ID"].nunique()
average_price = round(purchase_data["Price"].mean(),2)
purchases_count = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()
purchase_analysis = pd.DataFrame({"Number of Unique Items": [unique_items_count]
                                 ,"Average Price": [average_price]
                                 ,"Total Number of Purchases":[purchases_count]
                                 ,"Total Revenue":[total_revenue]})
purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_demo = purchase_data[['SN', 'Age', 'Gender']].groupby('SN').max()
male_count = gender_demo.loc[gender_demo["Gender"] == "Male",["Gender"]].count()
female_count = gender_demo.loc[gender_demo["Gender"] == "Female",["Gender"]].count()
other_count = gender_demo.loc[gender_demo["Gender"] == "Other / Non-Disclosed",["Gender"]].count()

total_gender_count = (male_count+female_count+other_count).max()
male_count = male_count.max()
female_count = female_count.max()
other_count = other_count.max()

summary_gender = pd.DataFrame.from_dict({"Gender": ["Male","Female","Other / Non-Disclosed"]
                                        ,"Total Count": [male_count,female_count,other_count]})

percent_players = [(round(male_count/total_gender_count,2)*100)
                                                        ,(round(female_count/total_gender_count,2)*100)
                                                        ,(round(other_count/total_gender_count,2)*100)]
summary_gender["Percentage of Players"]=percent_players

summary_gender = summary_gender.set_index("Gender")
summary_gender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
summary_df = purchase_data.copy()

summary_df = summary_df[["Gender","Item ID","Price","SN"]]
summary_df_pc = summary_df[["Gender","Item ID"]].groupby("Gender",as_index=False).count()
summary_df_pc = summary_df_pc.rename(columns={"Item ID":"Purchase Count"})
summary_df_app = round(summary_df[["Gender","Price"]].groupby("Gender",as_index=False).mean(),2)
summary_df_app = summary_df_app.rename(columns={"Price":"Average Purchase Price"})
summary_df_tpv = summary_df[["Gender","Price"]].groupby("Gender",as_index=False).sum()
summary_df_tpv = summary_df_tpv.rename(columns={"Price":"Total Purchase Value"})

summary_df_merge = summary_df_pc.merge(summary_df_app, on="Gender")
summary_df_merge = summary_df_merge.merge(summary_df_tpv, on="Gender")

summary_df_atppp = summary_df[["Gender","SN"]].groupby("Gender" ,as_index=False).nunique()
summary_df_atppp = summary_df_atppp.rename(columns={"SN":"Unique users per gender"})
summary_df_merge["Average Purchase Total per Person by Gender"] = round(summary_df_merge["Total Purchase Value"] / summary_df_atppp["Unique users per gender"],2)

summary_df_merge = summary_df_merge.set_index("Gender")
summary_df_merge
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0,9,14,19,24,29,34,39,200]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_analysis_age_df = purchase_data.copy()
purchase_analysis_age_df = purchase_analysis_age_df[["Age","SN","Item ID","Price"]]


purchase_analysis_age_df["Ages"] = pd.cut(purchase_analysis_age_df["Age"],bins=bins,labels=group_names)


purchase_analysis_age_df_tpv = purchase_analysis_age_df[["Ages","SN","Price"]].groupby("Ages" ,as_index=False).sum()
purchase_analysis_age_df_tpv = purchase_analysis_age_df_tpv.rename(columns={"Price":"Total Purchase Value"})

purchase_analysis_age_df_pc = purchase_analysis_age_df[["Ages","SN"]].groupby("Ages" ,as_index=False).count()
purchase_analysis_age_df_pc = purchase_analysis_age_df_pc.rename(columns={"SN":"Purchase Count"})

purchase_analysis_age_df_app = round(purchase_analysis_age_df[["Ages","Price"]].groupby("Ages" ,as_index=False).mean(),2)
purchase_analysis_age_df_app = purchase_analysis_age_df_app.rename(columns={"Price":"Average Purchase Price"})

purchase_analysis_age_df_users = purchase_analysis_age_df[["Ages","SN"]].groupby("Ages" ,as_index=False).nunique()

purchase_analysis_age_df_merged = purchase_analysis_age_df_tpv.merge(purchase_analysis_age_df_pc, on="Ages")
purchase_analysis_age_df_merged = purchase_analysis_age_df_merged.merge(purchase_analysis_age_df_app, on="Ages")

purchase_analysis_age_df_merged["Average Purchase Total per Person by Age Group"] = round(purchase_analysis_age_df_merged["Total Purchase Value"] / purchase_analysis_age_df_users["SN"],2)
purchase_analysis_age_df_merged = purchase_analysis_age_df_merged[["Ages","Purchase Count","Average Purchase Price","Total Purchase Value","Average Purchase Total per Person by Age Group"]]
# purchase_analysis_age_df_merged = purchase_analysis_age_df_merged.set_index("Ages")
purchase_analysis_age_df_merged


```

```python


age_demo = purchase_analysis_age_df_merged.copy()
age_demo["Total Count"] =  (purchase_analysis_age_df_users["SN"])
age_demo["Percentage of Players"] =  round((purchase_analysis_age_df_users["SN"] / (purchase_analysis_age_df_users["SN"].sum())) * 100,2)
age_demo = age_demo[["Ages","Total Count","Percentage of Players"]]
age_demo
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders_df = purchase_data.copy()
top_spenders_df_group_pc = top_spenders_df[["SN","Purchase ID"]].groupby("SN",as_index=False).count()
top_spenders_df_group_avp = round(top_spenders_df[["SN","Price"]].groupby("SN",as_index=False).mean(),2)
top_spenders_df_group_tpv = top_spenders_df[["SN","Price"]].groupby("SN",as_index=False).sum()
top_spenders_df_group_pc = top_spenders_df_group_pc.rename(columns={"Purchase ID":"Purchase Count"})
top_spenders_df_group_avp = top_spenders_df_group_avp.rename(columns={"Price":"Average Purchase Price"})
top_spenders_df_group_tpv = top_spenders_df_group_tpv.rename(columns={"Price":"Total Purchase Value"})
top_spenders_merged_df = top_spenders_df_group_pc.merge(top_spenders_df_group_avp, on="SN")
top_spenders_merged_df = top_spenders_merged_df.merge(top_spenders_df_group_tpv, on="SN")
top_spenders_merged_df = top_spenders_merged_df.sort_values("Total Purchase Value", ascending = False)
top_spenders_merged_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_pop_items_df = purchase_data.copy()
most_pop_items_df = most_pop_items_df[["Item ID","Item Name","Price"]]
most_pop_item_and_item_id = purchase_data.copy()
most_pop_item_and_item_id = most_pop_item_and_item_id[["Item ID","Item Name"]]#to add item id back
most_pop_item_and_item_id = most_pop_item_and_item_id.drop_duplicates()

most_pop_items_df_grp_tpv = most_pop_items_df[["Item ID","Item Name","Price"]].groupby("Item ID" and "Item Name",as_index=False).sum()
most_pop_items_df_grp_pc = most_pop_items_df[["Item ID","Item Name","Price"]].groupby("Item ID" and "Item Name",as_index=False).count()
most_pop_items_df_grp_ip = most_pop_items_df[["Item ID","Item Name","Price"]].groupby("Item ID" and "Item Name",as_index=False).max()

most_pop_items_df_grp_tpv = most_pop_items_df_grp_tpv[["Item Name","Price"]]
most_pop_items_df_grp_tpv = most_pop_items_df_grp_tpv.rename(columns={"Price":"Total Purchase Value"})

most_pop_items_df_grp_pc = most_pop_items_df_grp_pc[["Item Name","Item ID"]]
most_pop_items_df_grp_pc = most_pop_items_df_grp_pc.rename(columns={"Item ID":"Purchase Count"})

most_pop_items_df_grp_ip = most_pop_items_df_grp_ip[["Item Name","Price"]]
most_pop_items_df_merged = most_pop_items_df_grp_tpv.merge(most_pop_items_df_grp_pc, on = "Item Name")
most_pop_items_df_merged = most_pop_items_df_merged.merge(most_pop_items_df_grp_ip, on = "Item Name")
most_pop_items_df_merged = most_pop_items_df_merged.merge(most_pop_item_and_item_id, on = "Item Name")

most_pop_items_df_merged_sort_by_pc = most_pop_items_df_merged.sort_values("Purchase Count", ascending = False)

most_pop_items_df_merged_sort_by_pc = most_pop_items_df_merged_sort_by_pc[["Item ID","Item Name","Purchase Count","Price","Total Purchase Value"]]
most_pop_items_df_merged_sort_by_pc
```
